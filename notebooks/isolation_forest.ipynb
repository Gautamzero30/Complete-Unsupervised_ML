```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/thyroid_dataset.csv")
```

```python
df.head()
```

```python
X = df.drop("Outlier_label",axis
= 1)
```

```python
y=df["Outlier_label"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.ensemble import IsolationForest
clf= IsolationForest(
    n_estimators=200,contamination=0.036,random_state=42

) 
```

```python
labels = clf.fit_predict(X_scaled
)
```

```python
labels
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca= pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8,6))
```

```python
plt.scatter(
    x= X_pca[:,0],y=X_pca[:,1],
    c=labels
)
plt.xlabel("pc1")
plt.ylabel("pc2")
```

```python
import numpy as np
n_outliers = np.sum(labels==-1)
n_normal= np.sum(labels ==1)
print("outlier = ", n_outliers)
print("normal = ", n_normal)
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
neighbs = LocalOutlierFactor()
labels = neighbs.fit_predict(X_scaled)
```

```python
plt.figure(figsize= (8,6))
plt.scatter(X_pca[:,0],X_pca[:,1],c=labels)
```
